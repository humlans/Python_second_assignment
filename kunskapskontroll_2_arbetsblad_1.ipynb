# Remark on Broadcasting when doing Linear Algebra calculations in Python. 

### From the mathematical rules of matrix addition, the operation below (m1 + m2) does not make sense. The reason is that matrix addition requires two matrices of the same size. In Python however, it works due to broadcasting rules in NumPy. So you must be careful when doing Linear Algebra calculations in Python since they do not follow the "mathematical rules". This can however easily be handled by doing some simple programming, for example validating that two matrices have the same shape is easy if you for instance want to add two matrices. 

```python
m1 = np.array([[1, 2], [3, 4]])
m2 = np.array([1, 1])
print(m1 + m2)
```

### The example below would also not be allowed if following the "mathematical rules" in Linear Algebra. But it works due to broadcasting in NumPy. 

```python
v1 = np.array([1, 2, 3])
print(v1 + 1)
```

**Write a function "add_mult_matrices" that takes two matrices as input arguments (validate that the input are of the type numpy.ndarray by using the isinstance function), a third argument that is either 'add' or 'multiply' that specifies if you want to add or multiply the matrices (validate that the third argument is either 'add' or 'multiply'). When doing matrix addition, validate that the matrices have the same size. When doing matrix multiplication, validate that the sizes conform (i.e. number of columns in the first matrix is equal to the number of rows in the second matrix).**

**The function you create can be used for question 2 and 4 below.**

```python
def add_mult_matrices(matrix_1, matrix_2, add_or_multiply):
    """Function to either adds or multiplies two matrices using matrix multiplication"""
    if isinstance(matrix_1, np.ndarray) and isinstance(matrix_2, np.ndarray):
        if add_or_multiply == 'add':
            if matrix_1.shape == matrix_2.shape:
                return matrix_1 + matrix_2
            else:
                return('For matrix addition has to have the same dimensions.')
        elif add_or_multiply == 'multiply':
            if matrix_1.shape[1] == matrix_2.shape[0]:
                return matrix_1 @ matrix_2
            else:
                return('The first matrix columns and the second matrix rows has to match')
        else:
            raise Exception('The third value has to be either "add" or "multiply".')
    else:
        raise Exception('The arguments have to be of the type np.array')
    
```

### Copies and Views
Read the following link: https://numpy.org/doc/stable/user/basics.copies.html

**Basic indexing creates a view, How can you check if v1 and v2 is a view or copy? If you change the last element in v2 to 123, will the last element in v1 be changed? Why?**

```python
v1 = np.arange(4)
v2 = v1[-2:]
print(v1)
print(v2)
```

```python
# The base attribute of a view returns the original array while it returns None for a copy.
print(v1.base)
print(v2.base)
```

```python
# The last element in v1 will be changed aswell since v2 is a view, meaning they share the same data buffer.
v2[-1] = 123
print(v1)
print(v2)
```

#### Anwer

To check whether an array is a copy or a view you can use the base attribute, which will return None when the array is a copy. If you change an element in a view you also affect the original array that view is based on. That is why v1 is also changed when v2 is changed.

### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

##### Answer

The two ways to plot are:
1. "Object oriented style"
2. "Pyplot style"

The first way is the recommended approach because it is more flexible and more customizable when working with more data or more complex plots. To create a plot you first have to create a figure and how many axes this figure will contain. Then you can add labels, text, change the color and many other thing onto the axes until it looks the same as the image in your head.

The second way is faster but not as flexible. Using pyplot to create a plot works fine if it is a simple plot without much detail. The difference is that we use the methods from the matplotlib library to create the plot while in the object oriented approach we create an object and make changes to this object. To change title or a label in a pyplot-style created figure you would have to use the function .title() while in the object-oriented style you set a property which means that set_title() is used when changing and you can get it by only typing ax.title.

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

Figure: It is the "box" that contains all the other object, such as axes, axis, titles, colorbars etc. It it the topmost "layer" to the image and can also be liked to a canvas.
<br>Axes: A subplot of the figure in which the axis, plotlines and text is drawn.
<br>Axis: The lines to show the dimensions of the figure, there can be one, two or three. Also contains the ticks and their labels.
<br>Artist: Figure, text, axes, axis are all examples of artists. Artists are objects that gets painted onto the canvas.

### When plotting in Matplotlib, what is the expected input data type?

The expected input is numpy.array or numpy.ma.masked_array. It also accepts inputs that can be converted to numpy.arrays, for example lists or tuples.  

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
# Object-oriented approach

x = np.linspace(-4, 4, 100)
fig, ax = plt.subplots(figsize=(5, 4), layout='constrained')
ax.plot(x, x**2, label='$y = x^2$')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('A simple equation system')
ax.legend(loc='best')
print()
```

```python
# Pyplot approach

x = np.linspace(-4, 4, 1000)
plt.figure(figsize=(5, 4), layout='constrained')
plt.plot(x, x**2, label='$y = x^2$')
plt.xlabel('x')
plt.ylabel('y')
plt.title("A simple equation system")
plt.legend()
print()
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
np.random.seed(15)

# Data for scatter plot
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
fig, ax = plt.subplots(1, 2, layout='constrained')
fig.suptitle('Two different plots', size=20, weight='bold')

ax[0].scatter(random_data_x, random_data_y, color='orange', edgecolors= "black")
ax[0].set_title('Scatter plot')

ax[1].bar(names, values, color='red')
ax[1].set_title('Bar plot')
ax[1].set_ylabel('Number of fruits')
print()
```
